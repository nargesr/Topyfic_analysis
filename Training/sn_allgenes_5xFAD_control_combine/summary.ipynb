```python
#from scipy.interpolate import interp1d
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['pdf.fonttype'] = 42
sns.set_context('paper')

fig, ax = plt.subplots(figsize=(5,6), facecolor='white')

sns.scatterplot(data=df, x="k", y="N", s=100,
                hue="Genotype",
                palette={'5xFAD/BL6': '#FAA51A', 
                         'C57BL/6J': '#6ABD45', 
                         'Combined': 'darkgrey'})

#plt.plot(xnew, f_linear(xnew), '-', label='linear')
    
plt.plot([0] + k_topics + [55], 
         [0] + k_topics + [55], 
         linestyle='dashed',
         color='lightgrey')    
plt.axvline(x=15, color="salmon", linestyle="dashed")
plt.xlabel('K (#topics in each LDA run)')
plt.ylabel('N (#rLDA topics)')
plt.ylim([0,55])
plt.xlim([0,55])
plt.title('single nucleus RNA-seq data using All genes')
#plt.legend(loc='best')
plt.savefig(f'k_n.pdf')
```

```python
#from scipy.interpolate import interp1d
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['pdf.fonttype'] = 42

x = df['k'].tolist()
y = df['N'].tolist()

plt.plot(x, y, 'o')
    
# Define interpolators.
x = df['k'].values.reshape(-1, 1)
y = df['N'].values.reshape(-1, 1)
model = LinearRegression()
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)

x_pred = np.linspace(5, 50, num=41, endpoint=True)
y_pred = model.predict(x_pred.reshape(-1, 1))
plt.plot(x_pred, y_pred, 
             linestyle='dashed', 
             label=fr'R^2 = {round(r_sq, 5)}')

    
plt.plot([0] + k_topics, [0] + k_topics, linestyle='solid', label='x=y')    
#plt.axvline(x=10, color="red", linestyle="solid", label='k=10')
plt.xlabel('K (#topics in each LDA run)')
plt.ylabel('N (#rLDA topics)')
plt.ylim([0,50])
plt.xlim([0,50])
plt.title('single nucleus RNA-seq data from Model AD and ENCODE\n (WT mice, All genes)')
plt.legend(loc='best')
plt.savefig(f'k_n_prediction.pdf')
```
